# 1.職員最多的學校

```python
origin_data['總職員數'] = origin_data[['專任教師數男', '專任教師數女', '兼任教師數男', '兼任教師數女', '職員數男', '職員數女']].sum(axis=1) #在dataframe檔中加入並計算每所學校的總職員數
#axis=1橫向加總(一所學校中的不同職員總和) #axis=0縱向加總(每所學校中的同類職員總和)
worker_sum_in_order = origin_data.sort_values(by='總職員數', ascending=False) #設立新dataframe,以職員數降序排序(ascending=False)
# worker_sum_in_order[['學校名稱','總職員數']] #顯示每間學校的總職員數
print(f"最多職員的學校是:{worker_sum_in_order.head(1)['學校名稱'].iloc[0]},有{worker_sum_in_order.head(1)['總職員數'].iloc[0]}人") #顯示最多職員的學校(排第一者)
#iloc:不要印出那些colab自己顯示的無謂資訊
```

```python
school_name = worker_sum_in_order.head(6)['學校名稱'] #設置x軸(取前6筆)
worker_sum = worker_sum_in_order.head(6)['總職員數'] #設置y軸(取前6筆)
matplotlib.pyplot.bar(school_name, worker_sum, width=0.4, color='#93282c') #設置圖表
matplotlib.pyplot.title('各校職員人數') #圖表標題
matplotlib.pyplot.xlabel('校名') #x軸標題
matplotlib.pyplot.ylabel('職員數') #y軸標題
matplotlib.pyplot.show() #顯示
```

# 2.各縣市有幾所學校

```python
one_data_for_each_school = origin_data.drop_duplicates(subset=['學校名稱']) #建立新dataframe,但刪去"學校名稱"重複者
city_count = one_data_for_each_school['縣市名稱'].value_counts() #計算每個縣市有多少學校
city_list = list(origin_data['縣市名稱'].unique()) #建立縣市名的list
print(f"一共有 {len(one_data_for_each_school)} 所學校，其中：")
for i in city_list:
  print(f"位於{i}的學校共有{city_count[i]}所") #使縣市名list與其對應的學校數量配對
```

```python
grouped_by_city = origin_data.groupby("縣市名稱") #將原資料照"縣市名稱"分類
count_grouped_by_city = grouped_by_city["縣市名稱"].count() #計算每個分類有多少筆
count_grouped_by_city.plot(kind="pie", title="學校分布") #畫圖
matplotlib.pyplot.ylabel("") #設置y軸標籤為空
```

# 3.師生比

```python
origin_data['總老師數'] = origin_data[['專任教師數男', '專任教師數女', '兼任教師數男', '兼任教師數女']].sum(axis=1) #在dataframe中加入並計算每所學校的總老師數
#axis=1橫向加總(一所學校中的不同老師總和)
total_teachers = origin_data[['總老師數']].sum(axis=0) #將每所學校的老師加總 #axis=0縱向加總(每所學校中的同類老師總和)

origin_data['總學生數'] = origin_data[['學生數男','學生數女']].sum(axis=1) #在dataframe中加入並計算每所學校的總學生數
#axis=1橫向加總(一所學校中的不同學生總和)
total_students = origin_data[['總學生數']].sum(axis=0) #將每所學校的學生加總 #axis=0縱向加總(每所學校中的同類學生總和)

print(f'共有{total_teachers.iloc[0]}位老師，{total_students.iloc[0]}位學生') ##iloc:不要印出那些colab自己顯示的無謂資訊
students_per_teacher_float = total_students.values/total_teachers.values #學生老師相除得浮點數
import numpy #students_per_teacher_float型別為numpy.ndarray，需要導入numpy方可使用round(四捨五入)運算
students_per_teacher_int = numpy.ndarray.round(students_per_teacher_float,0) #round(變數名,小數點後幾位):四捨五入
print(f'平均每位老師對應{students_per_teacher_int}位學生')
```

```python
origin_data['師生比'] = origin_data['總學生數']/origin_data['總老師數'] #dataframe新增師生比
school_name = origin_data.head(6)['學校名稱'] #設置x軸(取前6筆)
teacher_student_ratio = origin_data.head(6)['師生比'] #設置y軸(取前6筆)
matplotlib.pyplot.bar(school_name, teacher_student_ratio, width=0.4, color='#93282c') #設置圖表
matplotlib.pyplot.title('各校師生比') #圖表標題
matplotlib.pyplot.xlabel('校名') #x軸標題
matplotlib.pyplot.ylabel('師生比') #y軸標題
matplotlib.pyplot.show() #顯示
```

# 4.最多女生的學校

```python
origin_data['總女生數'] = origin_data[['專任教師數女','兼任教師數女','職員數女','學生數女']].sum(axis=1) #在dataframe檔中加入並計算每所學校的總女生數
#axis=1橫向加總(一所學校中的不同女生總和) #axis=0縱向加總(每所學校中的同類女生總和)
women_sum_in_order = origin_data.sort_values(by='總女生數', ascending=False) #設立新dataframe,以女生數降序排序(ascending=False)
# women_sum_in_order[['學校名稱','總女生數']] #顯示每間學校的總女生數
print(f"最多女生的學校是:{women_sum_in_order.head(1)['學校名稱'].iloc[0]},有{women_sum_in_order.head(1)['總女生數'].iloc[0]}人") #顯示最多女生的學校(排第一者)
#iloc:不要印出那些colab自己顯示的無謂資訊
```

```python
school_name = women_sum_in_order.head(6)['學校名稱'] #設置x軸(取前6筆)
women_sum = women_sum_in_order.head(6)['總女生數'] #設置y軸(取前6筆)
matplotlib.pyplot.bar(school_name, women_sum, width=0.4, color='#55123c') #設置圖表
matplotlib.pyplot.title('各校女生人數') #圖表標題
matplotlib.pyplot.xlabel('校名') #x軸標題
matplotlib.pyplot.ylabel('女生數') #y軸標題
matplotlib.pyplot.show() #顯示
```
